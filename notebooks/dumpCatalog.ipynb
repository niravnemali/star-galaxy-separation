```python
def getMag(flux, fluxErr):
    ZP = 31.4  # zeropoint for AB mags
    mag = ZP - 2.5 * np.log10(flux)
    magErr = 1.0857 * (fluxErr / flux)
    return mag, magErr 
```

```python
objects['g_cModelMag'], objects['g_cModelMagErr'] = getMag(objects['g_cModelFlux'], objects['g_cModelFluxErr'])
objects['r_cModelMag'], objects['r_cModelMagErr'] = getMag(objects['r_cModelFlux'], objects['r_cModelFluxErr'])
```

```python
plt.plot(objects['g_cModelMag'] - objects['r_cModelMag'], objects['g_cModelMag'], 'o', ms=2, alpha=0.2, mew=0)
```

```python
C = objects['g_cModelMag'] - objects['r_cModelMag']
M = objects['g_cModelMag']
goodC = C[(M<27)&(M>16)&(C>-1)&(C<3)]
goodM = M[(M<27)&(M>16)&(C>-1)&(C<3)]
```

```python
plt.plot(goodC, -goodM, 'o', ms=2, alpha=0.2, mew=0)
```
